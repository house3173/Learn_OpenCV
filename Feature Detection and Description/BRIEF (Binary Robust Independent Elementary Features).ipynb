<h4>Phân tích</h4>

SIFT sử dụng một vector 128 chiều để biểu diễn descriptor, do sử dụng số thực dấu phẩy động nên về cơ bản nó cần 512 bytes. Tương tự SURF cũng sử dụng 256 bytes cho 64 chiều. Việc phải tạo các vector như vậy cho hàng nghìn keypoints cần rất nhiều bộ nhớ, làm cho thời gian so khớp lâu hơn và cũng không phù hợp với các hệ thống có tài nguyên nhỏ như hệ thống nhúng. Để thực hiện được thì cần phải sử dụng các bộ nén như PCA, LDA.. hay sử dụng hàm băm LSH để chuyển đổi các bộ mô tả này sang dạng nhị phân, các chuỗi nhị phân này khi so khớp sử dụng khoảng cách Hamming (tính toán đơn giản bởi dùng phép XOR). Tuy nhiên, việc cần lưu trữ lượng lớn thông tin cho bộ mô tả như vậy là dư thừa, chiếm dụng tài nguyên mặc dù có phương pháp tăng tốc độ so khớp.

BRIEF được sử dụng để mô tả các điểm đặc trưng bằng một chuỗi nhị phân mà không cần phải xác định các descriptor. Về cơ bản, BRIEF tạo ra một bộ mẫu gồm các cặp điểm ngẫu nhiên xung quanh mỗi điểm đặc trưng. Các cặp điểm này thường được chọn trong một vùng lân cận hình vuông (patch) có kích thước cố định (ví dụ 31x31 pixel) quanh điểm đặc trưng. Đối với mỗi cặp điểm (p, q) trong bộ mẫu, so sánh cường độ (giá trị màu xám) của hai điểm. Nếu cường độ tại p lớn hơn cường độ tại q, gán giá trị 1, ngược lại gán giá trị 0. Kết quả của mỗi so sánh là một bit trong mô tả nhị phân. Tập hợp các bit từ các so sánh cường độ thành một chuỗi nhị phân. Chuỗi này chính là mô tả đặc trưng BRIEF của điểm đặc trưng đó.

Số lượng cặp nhị phân được chọn thường là 128, 256, 512 bits (theo mặc định là 256 bits). OpenCV biểu thị bằng bytes nên nó sẽ là 16, 32, 64. 

Đặc biệt, BRIEF chỉ là một phương pháp để tạo bộ mô tả tính năng chứ không cung cấp thuật toán hỗ trợ phát hiện các tính năng. Vì vậy, trước tiên cần sử dụng các bộ phát hiện tính năng như SIFT, SURF, FAST... để tìm các keypoints, sau đó mới áp dụng BRIEF để tạo bộ mô tả descriptor và cuối cùng dùng khoảng cách Hamming (số lượng bit khác nhau giữa hai chuỗi bit cùng độ dài, dùng phép XOR) để so khớp tính năng. Bổ sung, phương pháp phát hiện tính năng CenSurE được khuyến cáo sử dụng để tìm các keypoints phục vụ cho BRIEF.

Hạn chế của BRIEF: BRIEF có thể nhạy cảm với nhiễu và các biến đổi hình học phức tạp như co giãn hoặc biến dạng phi tuyến. BRIEF không bất biến với sự quay, do đó nó có thể không phù hợp trong các tình huống mà đối tượng có thể xoay nhiều trong không gian hình ảnh.

<h4>BRIEF với FAST, SIFT</h4>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Đọc hình ảnh
img = cv2.imread('sift.png', 0)

# Sử dụng SURF detector
fast = cv2.FastFeatureDetector_create(8, True, cv2.FAST_FEATURE_DETECTOR_TYPE_9_16)
keypoints_fast = fast.detect(img, None)

# Sử dụng SIFT detector
sift = cv2.SIFT_create()
keypoints_sift = sift.detect(img, None)

# Khởi tạo BRIEF extractor
brief_fast = cv2.xfeatures2d.BriefDescriptorExtractor_create()
brief_sift = cv2.xfeatures2d.BriefDescriptorExtractor_create()

# Tính toán mô tả BRIEF cho các điểm đặc trưng
keypoints_fast, descriptors_fast = brief_fast.compute(img, keypoints_fast)
keypoints_sift, descriptors_sift = brief_sift.compute(img, keypoints_sift)

# Vẽ các điểm đặc trưng trên hình ảnh
img_keypoints_fast = cv2.drawKeypoints(img, keypoints_fast, None, color=(0, 255, 0))
img_keypoints_sift = cv2.drawKeypoints(img, keypoints_sift, None, color=(0, 255, 0))

# Hiển thị hình ảnh với các điểm đặc trưng
plt.figure(figsize=[12,6])
plt.subplot(121); plt.imshow(img_keypoints_fast, cmap='gray')
plt.title('FAST keypoints with BRIEF descriptors')

plt.subplot(122); plt.imshow(img_keypoints_sift, cmap='gray')
plt.title('SIFT keypoints with BRIEF descriptors')
plt.show()

# In một số mô tả đặc trưng BRIEF
print("Số lượng điểm đặc trưng: ", len(keypoints_fast))
print("Kích thước của mô tả đặc trưng: ", descriptors_fast.shape)
print("Mô tả đặc trưng BRIEF:\n", descriptors_fast[:5])
print("---------------------------------------------------------------")
print("SIFT keypoints with BRIEF descriptors")
print("Số lượng điểm đặc trưng: ", len(keypoints_sift))
print("Kích thước của mô tả đặc trưng: ", descriptors_sift.shape)
print("Mô tả đặc trưng BRIEF:\n", descriptors_sift[:5])
```

```python
# Khởi tạo bộ đối sánh BFMatcher
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

# Thực hiện đối sánh các mô tả điểm đặc trưng
matches = bf.match(descriptors_fast, descriptors_sift)

# Sắp xếp các đối sánh theo khoảng cách
matches = sorted(matches, key=lambda x: x.distance)

# Vẽ các đối sánh
img_matches = cv2.drawMatches(img, keypoints_fast, img, keypoints_sift, matches[:150], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Hiển thị ảnh với các đối sánh
plt.figure(figsize=(15,15))
plt.imshow(img_matches)
plt.title('BRIEF Keypoint Matches')
plt.show()
```

<h4>BRIEF với CenSurE</h4>

<h5>Thuật toán CenSurE</h5>

Các bước chính của CenSurE:

1. Gaussian Pyramid: CenSurE bắt đầu bằng cách tạo một Gaussian pyramid từ hình ảnh gốc. Gaussian pyramid là một cấu trúc đa tỉ lệ của hình ảnh, trong đó mỗi cấp độ là phiên bản làm mờ và giảm kích thước của cấp độ trước đó. Điều này giúp phát hiện các đặc trưng ở các tỉ lệ khác nhau.

2. Center-Surround Filters: Ở mỗi cấp độ của Gaussian pyramid, CenSurE áp dụng một loạt các bộ lọc Center-Surround. Các bộ lọc này tương tự như các bộ lọc được sử dụng trong thuật toán SIFT và DoG (Difference of Gaussians), nhưng CenSurE sử dụng các bộ lọc hình học khác nhau, chẳng hạn như hình tròn và hình chữ thập (cross-shaped filters). Các bộ lọc Center-Surround được tạo ra bằng cách xác định một vùng trung tâm và vùng xung quanh (ví dụ vùng trung tâm 3x3 và vùng xung quanh 5x5). Bộ lọc sẽ tính toán sự khác biệt giữa giá trị pixel của vùng trung tâm và vùng xung quanh tại mỗi điểm trong hình ảnh. Sau khi áp dụng bộ lọc, CenSurE tìm các điểm cực trị cục bộ, tức là các điểm mà sự khác biệt giữa vùng trung tâm và vùng xung quanh lớn nhất hoặc nhỏ nhất. Các điểm cực trị này là các điểm đặc trưng (keypoints) trong hình ảnh. Các bộ lọc này giúp xác định các đặc trưng có thể có ở các tỉ lệ và hướng khác nhau.

3. Tìm các điểm cực trị (Extrema): Sau khi áp dụng các bộ lọc Center-Surround, CenSurE tìm các điểm cực trị cục bộ (local extrema) trong các hình ảnh kết quả. Một điểm cực trị là điểm mà giá trị của nó lớn hơn hoặc nhỏ hơn các điểm lân cận trong không gian (x, y) và tỉ lệ (scale). Điều này có nghĩa là CenSurE tìm các điểm có cường độ sáng cao hoặc thấp hơn các điểm xung quanh trong không gian và tỉ lệ.

4. Kiểm tra ổn định và loại bỏ các điểm tương phản thấp: Các điểm cực trị được kiểm tra về độ ổn định và loại bỏ các điểm tương phản thấp hoặc nhiễu. Quá trình này có thể bao gồm kiểm tra giá trị của các điểm trong Gaussian pyramid và loại bỏ các điểm có độ tương phản thấp hoặc không có ý nghĩa.

5. Gán hướng cho các điểm đặc trưng: Cuối cùng, CenSurE gán hướng cho các điểm đặc trưng bằng cách tính toán gradient tại mỗi điểm. Hướng của gradient được sử dụng để làm cho các đặc trưng không phụ thuộc vào sự thay đổi hướng của hình ảnh.

<h5>Sử dụng BRIEF kết hợp CenSurE</h5>

```python
# Trong OpenCV, CenSurE detector được dùng với tên gọi STAR detector
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Đọc hình ảnh
img = cv2.imread('sift.png', 0)

# Khởi tạo CenSurE detector (STAR)
star = cv2.xfeatures2d.StarDetector_create()

# Tìm các điểm đặc trưng bằng STAR
keypoints = star.detect(img, None)

# Khởi tạo BRIEF extractor
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()

# Tính toán mô tả BRIEF cho các điểm đặc trưng
keypoints, descriptors = brief.compute(img, keypoints)

# Vẽ các điểm đặc trưng trên hình ảnh
img_keypoints = cv2.drawKeypoints(img, keypoints, None, color=(0, 255, 0))

# Hiển thị hình ảnh với các điểm đặc trưng
plt.imshow(img_keypoints, cmap='gray')
plt.title('CenSurE keypoints with BRIEF descriptors')
plt.show()

# In một số mô tả đặc trưng BRIEF để xem
print("Số lượng điểm đặc trưng: ", len(keypoints))
print("Kích thước của mô tả đặc trưng: ", descriptors.shape)
print("Mô tả đặc trưng BRIEF:\n", descriptors[:5])
```
