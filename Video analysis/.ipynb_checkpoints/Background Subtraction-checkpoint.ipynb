Background Subtraction (BS) l√† m·ªôt k·ªπ thu·∫≠t ph·ªï bi·∫øn v√† ƒë∆∞·ª£c s·ª≠ d·ª•ng r·ªông r√£i ƒë·ªÉ t·∫°o foreground mask (c·ª• th·ªÉ l√† h√¨nh ·∫£nh nh·ªã ph√¢n ch·ª©a c√°c ƒëi·ªÉm ·∫£nh thu·ªôc v·ªÅ c√°c ƒë·ªëi t∆∞·ª£ng chuy·ªÉn ƒë·ªông trong c·∫£nh) b·∫±ng c√°ch s·ª≠ d·ª•ng static camera. BS t√≠nh to√°n foreground mask b·∫±ng c√°ch th·ª±c hi·ªán ph√©p tr·ª´ gi·ªØa khung h√¨nh hi·ªán t·∫°i v√† khung h√¨nh n·ªÅn, ch·ª©a ph·∫ßn tƒ©nh c·ªßa c·∫£nh ho·∫∑c n√≥i chung l√† m·ªçi th·ª© c√≥ th·ªÉ ƒë∆∞·ª£c coi l√† n·ªÅn d·ª±a tr√™n c√°c ƒë·∫∑c ƒëi·ªÉm c·ªßa c·∫£nh ƒë∆∞·ª£c quan s√°t.

<h5>1. BackgroundSubtractorMOG2 (Improved Mixture of Gaussians)</h5>

Thu·∫≠t to√°n MOG (Mixture of Gaussians) s·ª≠ d·ª•ng m·ªôt ph∆∞∆°ng ph√°p ƒë·ªÉ m√¥ h√¨nh h√≥a m·ªói pixel c·ªßa n·ªÅn b·∫±ng m·ªôt h·ªón h·ª£p c·ªßa ùêæ ph√¢n ph·ªëi Gaussian (th∆∞·ªùng 
ùêæ t·ª´ 3 ƒë·∫øn 5). Tr·ªçng s·ªë c·ªßa c√°c ph√¢n ph·ªëi Gaussian trong h·ªón h·ª£p n√†y bi·ªÉu th·ªã t·ª∑ l·ªá th·ªùi gian m√† c√°c m√†u s·∫Øc ƒë√≥ xu·∫•t hi·ªán trong c·∫£nh. C√°c m√†u s·∫Øc c√≥ kh·∫£ nƒÉng l√† n·ªÅn th∆∞·ªùng l√† nh·ªØng m√†u t·ªìn t·∫°i l√¢u h∆°n v√† √≠t thay ƒë·ªïi h∆°n.

ƒêi·ªÅu n√†y c√≥ nghƒ©a l√† n·∫øu m·ªôt m√†u s·∫Øc t·∫°i m·ªôt pixel xu·∫•t hi·ªán li√™n t·ª•c trong m·ªôt kho·∫£ng th·ªùi gian d√†i v√† √≠t thay ƒë·ªïi, th√¨ m√†u s·∫Øc ƒë√≥ c√≥ kh·∫£ nƒÉng l√† m√†u c·ªßa n·ªÅn. Ng∆∞·ª£c l·∫°i, c√°c m√†u s·∫Øc thay ƒë·ªïi nhanh ch√≥ng v√† kh√¥ng xu·∫•t hi·ªán th∆∞·ªùng xuy√™n s·∫Ω ƒë∆∞·ª£c coi l√† c√°c ƒë·ªëi t∆∞·ª£ng ti·ªÅn c·∫£nh di chuy·ªÉn.

C·ª• th·ªÉ h∆°n:
1. H·ªón h·ª£p c√°c ph√¢n ph·ªëi Gaussian: M·ªói pixel ƒë∆∞·ª£c m√¥ h√¨nh h√≥a b·∫±ng m·ªôt t·∫≠p h·ª£p ùêæ ph√¢n ph·ªëi Gaussian. M·ªói ph√¢n ph·ªëi Gaussian m√¥ t·∫£ m·ªôt "m√†u s·∫Øc" c·ª• th·ªÉ m√† pixel ƒë√≥ c√≥ th·ªÉ c√≥.
2. Tr·ªçng s·ªë: Tr·ªçng s·ªë c·ªßa m·ªói Gaussian bi·ªÉu th·ªã t·ª∑ l·ªá th·ªùi gian m√† m√†u s·∫Øc t∆∞∆°ng ·ª©ng xu·∫•t hi·ªán t·∫°i pixel ƒë√≥. C√°c tr·ªçng s·ªë n√†y ƒë∆∞·ª£c c·∫≠p nh·∫≠t li√™n t·ª•c khi c√≥ c√°c khung h√¨nh m·ªõi.
3. M√†u n·ªÅn: M√†u s·∫Øc n√†o c√≥ tr·ªçng s·ªë l·ªõn h∆°n, t·ª©c l√† xu·∫•t hi·ªán l√¢u h∆°n v√† √≠t thay ƒë·ªïi h∆°n, s·∫Ω ƒë∆∞·ª£c coi l√† m√†u n·ªÅn. M√†u s·∫Øc c√≥ tr·ªçng s·ªë th·∫•p h∆°n v√† thay ƒë·ªïi nhanh ch√≥ng ƒë∆∞·ª£c coi l√† ti·ªÅn c·∫£nh.

```python
import cv2
import matplotlib.pyplot as plt

cap = cv2.VideoCapture('vtest.avi')
fgbg = cv2.createBackgroundSubtractorMOG2()

count = 0
while True:
    count += 1
    ret, frame = cap.read()
    if not ret:
        break
    fgmask = fgbg.apply(frame)
    cv2.imshow('frame', fgmask)

    if count == 250:
        plt.figure(figsize=[12,6])
        plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB));
        plt.subplot(1,2,2); plt.imshow(fgmask, cmap='gray')
        plt.show()
        break
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

<h5>2. BackgroundSubtractorKNN (K-Nearest Neighbors)</h5>

Ph∆∞∆°ng ph√°p n√†y s·ª≠ d·ª•ng thu·∫≠t to√°n k-nearest neighbors ƒë·ªÉ ph√¢n lo·∫°i c√°c pixel.

Nguy√™n l√Ω ho·∫°t ƒë·ªông
1. Kh·ªüi t·∫°o: M·ªói pixel trong h√¨nh ·∫£nh ƒë∆∞·ª£c l∆∞u tr·ªØ d∆∞·ªõi d·∫°ng m·ªôt danh s√°ch c√°c gi√° tr·ªã pixel t·ª´ ùëÅ khung h√¨nh g·∫ßn ƒë√¢y nh·∫•t (v√≠ d·ª•: 50 khung h√¨nh). C√°c gi√° tr·ªã n√†y ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ ∆∞·ªõc t√≠nh n·ªÅn c·ªßa pixel ƒë√≥.
2. C·∫≠p nh·∫≠t n·ªÅn: Khi m·ªôt khung h√¨nh m·ªõi ƒë∆∞·ª£c ƒë·ªçc, gi√° tr·ªã c·ªßa m·ªói pixel ƒë∆∞·ª£c so s√°nh v·ªõi c√°c gi√° tr·ªã trong danh s√°ch l∆∞u tr·ªØ c·ªßa n√≥. Thu·∫≠t to√°n KNN ki·ªÉm tra gi√° tr·ªã pixel hi·ªán t·∫°i v·ªõi ùëò gi√° tr·ªã g·∫ßn nh·∫•t trong danh s√°ch l∆∞u tr·ªØ ƒë·ªÉ quy·∫øt ƒë·ªãnh li·ªáu pixel hi·ªán t·∫°i c√≥ thu·ªôc v·ªÅ n·ªÅn hay ti·ªÅn c·∫£nh.
3. Ph√¢n lo·∫°i n·ªÅn v√† ti·ªÅn c·∫£nh: M·ªôt pixel ƒë∆∞·ª£c coi l√† thu·ªôc v·ªÅ n·ªÅn n·∫øu c√≥ ƒë·ªß s·ªë l∆∞·ª£ng gi√° tr·ªã l√¢n c·∫≠n g·∫ßn nh·∫•t trong danh s√°ch l∆∞u tr·ªØ m√† t∆∞∆°ng ƒë·ªìng v·ªõi gi√° tr·ªã pixel hi·ªán t·∫°i (theo m·ªôt ng∆∞·ª°ng x√°c ƒë·ªãnh). Ng∆∞·ª£c l·∫°i, n·∫øu kh√¥ng ƒë·ªß s·ªë l∆∞·ª£ng gi√° tr·ªã l√¢n c·∫≠n t∆∞∆°ng ƒë·ªìng, pixel ƒë√≥ ƒë∆∞·ª£c coi l√† thu·ªôc v·ªÅ ti·ªÅn c·∫£nh.
4. C·∫≠p nh·∫≠t danh s√°ch l∆∞u tr·ªØ: N·∫øu m·ªôt pixel ƒë∆∞·ª£c coi l√† thu·ªôc v·ªÅ n·ªÅn, gi√° tr·ªã hi·ªán t·∫°i c·ªßa n√≥ ƒë∆∞·ª£c th√™m v√†o danh s√°ch l∆∞u tr·ªØ. ƒê·ªÉ tr√°nh tr√†n b·ªô nh·ªõ, m·ªôt gi√° tr·ªã c≈© trong danh s√°ch s·∫Ω b·ªã thay th·∫ø b·∫±ng gi√° tr·ªã m·ªõi n√†y.

```python
import cv2
import matplotlib.pyplot as plt

cap = cv2.VideoCapture('vtest.avi')
fgbg = cv2.createBackgroundSubtractorKNN()

count = 0
while True:
    count += 1
    ret, frame = cap.read()
    if not ret:
        break
    fgmask = fgbg.apply(frame)
    cv2.imshow('frame', fgmask)

    if count == 250:
        plt.figure(figsize=[12,6])
        plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB));
        plt.subplot(1,2,2); plt.imshow(fgmask, cmap='gray')
        plt.show()
        break
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

<h5>3. BackgroundSubtractorGMG (Godbehere-Matsukawa-Goldberg)</h5>

Ph∆∞∆°ng ph√°p n√†y s·ª≠ d·ª•ng m√¥ h√¨nh Bayes ƒë·ªÉ ph√¢n lo·∫°i c√°c pixel v√†o n·ªÅn ho·∫∑c ti·ªÅn c·∫£nh
