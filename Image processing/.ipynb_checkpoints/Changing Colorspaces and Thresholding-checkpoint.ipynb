<h3>1. Changing Colorspaces</h3>

```python
# Changing Color-space
# For color conversion, we use the function cv2.cvtColor(input_image, flag) where flag determines the type of conversion.
# For BGR -> Gray conversion we use the flags cv2.COLOR_BGR2GRAY. 
# For BGR -> HSV, we use the flag cv2.COLOR_BGR2HSV. 

import cv2
import matplotlib.pyplot as plt

image = cv2.imread('Hau.jpg')

image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=[15,5])

plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.imshow(cv2.cvtColor(image_hsv, cv2.COLOR_BGR2RGB))

plt.subplot(133)
plt.imshow(cv2.cvtColor(image_gray, cv2.COLOR_BGR2RGB))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('blue_image.jpg')
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# nh ngha phm vi mu xanh lam trong khng gian mu HSV
# Hue khong t 110-130, Saturation v Value c th ty chnh theo nhu cu
lower_blue = np.array([110, 50, 50])
upper_blue = np.array([130, 255, 255])

# To mt n  ch cha cc vng mu xanh lam ca image_hsv
mask = cv2.inRange(image_hsv, lower_blue, upper_blue)

# p dng mt n ln hnh nh gc  ch hin th khu vc mu xanh lam
blue_regions = cv2.bitwise_and(image, image, mask=mask)

# Hin th nh gc v nh ch cha vng mu xanh lam
plt.figure(figsize=[15, 5])

# Hin th nh gc
plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

# Hin th mt n
plt.subplot(132)
plt.imshow(mask, cmap='gray')
plt.title("Mask for Blue Regions")

# Hin th nh vi ch vng mu xanh lam
plt.subplot(133)
plt.imshow(cv2.cvtColor(blue_regions, cv2.COLOR_BGR2RGB))
plt.title("Blue Regions in Image")

plt.show()
```

```python
# xc nh phm vi mu trong khng gian HSV
green = np.uint8([[[0, 255, 0]]])
hsv_green = cv2.cvtColor(green, cv2.COLOR_BGR2HSV)
print(hsv_green)

# ly [H-10, 100,100] v [H+10, 255, 255] nh gii hn di v gii hn trn
```

<h3>2. Image Thresholding</h3>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('thresh_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

<h4>Ngng nh phn</h4h4>

```python
# if src(x,y) > thresh -->  dst(x,y) = maxValue
# else -->  dst(x,y) = 0

thresh_0 = 0
maxValue = 255
_, dst_0_255 = cv2.threshold(image, thresh_0, maxValue, cv2.THRESH_BINARY)

thresh_128 = 128
_, dst_128_255 = cv2.threshold(image, thresh_128, maxValue, cv2.THRESH_BINARY)

plt.figure(figsize=[15, 5])

plt.subplot(131)
plt.imshow(image, cmap='gray')
plt.subplot(132)
plt.imshow(dst_0_255, cmap='gray')
plt.subplot(133)
plt.imshow(dst_128_255, cmap='gray')

plt.show()
```

<h4>Ngng nh phn nghch o</h4>

```python
thresh = 128
maxValue = 255
th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_BINARY_INV)
plt.imshow(dst, cmap='gray')
```

<h4>Ngng ct bt ( THRESH_TRUNC )</h4>

```python
# if src(x,y) > thresh
#   dst(x,y) = thresh
# else
#   dst(x,y) = src(x,y)
# Trn ngng th a v ngng, di ngng th gi nguyn

thresh = 127
maxValue = 255
th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_TRUNC)
print(th)
plt.imshow(dst, cmap='gray')
```

<h4>Ngng v 0 ( THRESH_TOZERO )</h4>

```python
# if src(x,y) > thresh
#   dst(x,y) = src(x,y)
# else
#   dst(x,y) = 0
# Trn ngng th gi nguyn, di ngng th v 0

thresh = 127
maxValue = 255
th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_TOZERO)
plt.imshow(dst, cmap='gray')
```

<h4>Ngng o ngc v 0 ( THRESH_TOZERO_INV )</h4>

```python
# if src(x,y) > thresh
#   dst(x,y) = 0
# else
#   dst(x,y) = src(x,y)
# Trn ngng th v 0, di ngng th gi nguyn

thresh = 127
maxValue = 255
th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_TOZERO_INV)
plt.imshow(dst, cmap='gray')
```

<h4>Ngng thch ng</h4>

```python
# Thut ton tnh ngng s tnh gi tr ngng cho mi pixel hoc mi vng nh, tc l trong cng mt nh s c gi tr ngng khc nhau cho 
# cc vng/pixel khc nhau. V n mang li hiu qu tt cho nhng nh c nhiu  sng khc nhau.
#
# Adaptive Method - It decides how thresholding value is calculated.#     --> cv2.ADAPTIVE_THRESH_MEAN_C : threshold value is the mean of neighbourhood area.    
#     --> cv2.ADAPTIVE_THRESH_GAUSSIAN_C : threshold value is the weighted sum of neighbourhood values where weights are a gaussian window
## Block Size - It decides the size of neighbourhood area # 
# C - It is just a constant which is subtracted from the mean or weighted mean caculate

import cv2
import matplotlib.pyplot as plt

image = cv2.imread('adaptive_thresh.jpg', cv2.IMREAD_GRAYSCALE)

blockSize = 11
C = 2

image_median = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize, C)
image_gauss = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize, C)

plt.figure(figsize=[15, 5])

plt.subplot(131)
plt.imshow(image, cmap='gray')
plt.subplot(132)
plt.imshow(image_median, cmap='gray')
plt.subplot(133)
plt.imshow(image_gauss, cmap='gray')

plt.show()
```

<h4>Ngng Otsu</h4>

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('thresh_image.jpg', cv2.IMREAD_GRAYSCALE)

# p dng ngng Otsu
# Lu  rng nn truyn gi tr ngng ban u bng 0
# v ch nh loi phn ngng bng cv2.THRESH_BINARY + cv2.THRESH_OTSU
_, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=[10, 5])
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.subplot(122)
plt.imshow(otsu_thresh, cmap='gray')
plt.show()
```
