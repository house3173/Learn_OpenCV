<h4>L√Ω thuy·∫øt c·ªßa Harris Corner Detector (HCD)</h4>

<div  style="z-index: 2"><p>G√≥c l√† c√°c v√πng trong ·∫£nh c√≥ s·ª± thay ƒë·ªïi l·ªõn v·ªÅ c∆∞·ªùng ƒë·ªô theo m·ªçi h∆∞·ªõng. HCD d·ª±a v√†o s·ª± bi·∫øn ƒë·ªïi c∆∞·ªùng ƒë·ªô s√°ng m√† ph√°t hi·ªán ƒë·∫∑c tr∆∞ng g√≥c, ta c√≥ m·ªôt c·ª≠a s·ªï tr∆∞·ª£t tr√™n ·∫£nh khi c·ª≠a s·ªï ph√°t hi·ªán ra s·ª± thay ƒë·ªïi c∆∞·ªùng ƒë·ªô s√°ng ·ªü m·ªçi h∆∞·ªõng v√† s·ª± thay ƒë·ªïi ƒë√≥ l·ªõn h∆°n m·ªôt ng∆∞·ª°ng nh·∫•t ƒë·ªãnh th√¨ ƒë√≥ l√† g√≥c. </p></div>

Khi d·ªãch chuy·ªÉn m·ªôt c·ª≠a s·ªï nh·ªè qua c√°c v√πng c·ªßa ·∫£nh, ta th·∫•y:
<ul>
    <li><b> M·∫∑t ph·∫≥ng </b> : c·ª≠a s·ªï kh√¥ng c√≥ s·ª± thay ƒë·ªïi v·ªÅ c∆∞·ªùng ƒë·ªô khi di chuy·ªÉn theo b·∫•t k√¨ h∆∞·ªõng n√†o. </li>
    <li><b> C·∫°nh </b> : c·ª≠a s·ªï ch·ªâ thay ƒë·ªïi khi di chuy·ªÉn theo chi·ªÅu ngang </li>
    <li><b> G√≥c </b> : c·ª≠a s·ªï thay ƒë·ªïi khi di chuy·ªÉn b·∫•t k√¨ h∆∞·ªõng n√†o </li>
</ul>


Ta c√≥ c√¥ng th·ª©c ƒë·ªÉ t√≠nh l∆∞·ª£ng c∆∞·ªùng ƒë·ªô s√°ng thay ƒë·ªïi khi d·ªãch chuy·ªÉn c·ª≠a s·ªï 1 ƒëo·∫°n [u,v] nh∆∞ sau: <br>
$ \mathrm{E}(u,v) =  \displaystyle\sum_{x,y}^{} \mathrm{w}(x,y)
[\mathrm{I}(x+u,y+v)-\mathrm{I}(x,y)]2 $2<br>
Trong ƒë√≥:
<ul>
    <li><b>w(x,y)</b>: c·ª≠a s·ªï tr∆∞·ª£t t·∫°i t·ªça ƒë·ªô (x,y)</li>
    <li><b>I(x+u,y+v)</b>: c∆∞·ªùng ƒë·ªô t·∫°i t·ªça ƒë·ªô ƒë√£ d·ªãch chuy·ªÉn m·ªôt kho·∫£ng (u,v)</li>
    <li><b>I(x,y)</b>: c∆∞·ªùng ƒë·ªô t·∫°i t·ªça ƒë·ªô ƒëi·ªÉm hi·ªán t·∫°i</li>
    <li><b>E(u,v)</b>: s·ª± thay ƒë·ªïi c∆∞·ªùng ƒë·ªô v·ªõi c·ª≠a s·ªï (x,y) so v·ªõi c∆∞·ªùng ƒë·ªô v·ªõi c·ª≠a s·ªï (x+u, y+v)</li>
</ul> 

Tuy nhi√™n c√¥ng th·ª©c tr√™n t·ªën nhi·ªÅu th·ªùi gian ƒë·ªÉ t√≠nh n√™n ta s·∫Ω c√≥ th√™m v√†i b∆∞·ªõc bi·∫øn ƒë·ªïi (s·ª≠ d·ª•ng ph√©p bi·∫øn ƒë·ªïi Taylor):<br>
<img style="margin-left: 200px; height: 300px" src="https://images.viblo.asia/0b6ffc3e-2524-459a-9940-e7ad9e6ba321.png">

----> $\mathrm{E}(u,v) \approx \begin{bmatrix}\ u & \ v\\\end{bmatrix}\displaystyle\sum_{x,y}^{} \mathrm{w}(x,y)\begin{bmatrix}\ \mathrm{I_x^2}(x,y) & \ \mathrm{I_{xy}}(x,y)\\\ \mathrm{I_{xy}}(x,y) & \mathrm{I_y^2}()x,y) \\\end{bmatrix}\begin{bmatrix}\ u\\\ v \\\end{bmatrix} $e<br>
----> $\mathrm{M} =  \displaystyle\sum_{x,y}^{} \mathrm{w}(x,y)\begin{bmatrix}\ \mathrm{I_x^2}(x,y) & \ \mathrm{I_{xy}}(x,y)\\\ \mathrm{I_{xy}}(x,y) & \mathrm{I_y^2}(x,y) \\\end{bmatrix}$
<br>

M ·ªü ƒë√¢y l√† m·ªôt structure tensor hay ƒë∆∞·ª£c bi·∫øt ƒë·∫øn l√† second-moment matrix ƒë∆∞·ª£c c·∫•u th√†nh t·ª´ ùêºùë• v√† ùêºùë¶ (ƒë·∫°o h√†m theo h∆∞·ªõng x v√† y c·ªßa ·∫£nh), n√≥ cho ta bi·∫øt ph√¢n ph·ªëi ƒë·∫°o h√†m xung quanh m·ªôt ƒëi·ªÉm ƒë·ªÉ t·ª´ ƒë√≥ c√≥ th·ªÉ nh·∫≠n bi·∫øt ƒë∆∞·ª£c ƒë√≥ l√† c·∫°nh, g√≥c hay m·∫∑t ph·∫≥ng.
<br>

Vi·ªác ph√¢n lo·∫°i v√† t√¨m ra c√°c ƒëi·ªÉm ·∫£nh c√≥ ph·∫£i l√† c√°c g√≥c hay kh√¥ng d·ª±a v√†o vi·ªác t√≠nh gi√° tr·ªã R ( corner score function ). C√¥ng th·ª©c R nh∆∞ sau:<br>
$\mathrm{R} =  \mathrm{det(M)} - k(\mathrm{trace(M)})^2\hspace{1cm}$
<br>
Trong ƒë√≥:
<ul>
    <li>det(M) = $\lambda _{1}\lambda _{2}$</li>
    <li>trace(M) = $\lambda _{1} + \lambda _{2}$</li>
    <li>ùëò l√† h·∫±ng s·ªë trong ƒëo·∫°n [0.04, 0.06]</li>
</ul>

Khi ƒë√£ c√≥ ƒë∆∞·ª£c gi√° tr·ªã R, ch√∫ng ta s·∫Ω x√°c ƒë·ªãnh xem v√πng pixel n√†y c√≥ ph·∫£i g√≥c hay kh√¥ng.
<ul>
    <li>N·∫øu gi√° tr·ªã |R| nh·ªè, khi ùúÜ1 v√† ùúÜ2 ƒë·ªÅu nh·ªè th√¨ v√πng pixel ƒë√≥ l√† m·∫∑t ph·∫≥ng.</li>
    <li>N·∫øu R < 0, t·ª©c l√† khi ùúÜ1 >> ùúÜ2 ho·∫∑c ùúÜ2 >> ùúÜ1, th√¨ v√πng pixel ƒë√≥ l√† c·∫°nh.</li>
    <li>N·∫øu gi√° tr·ªã R > 0, t·ª©c l√† khi ùúÜ1 >> 1, ùúÜ2 >> 1 v√† ùúÜ1 ‚âÉ ùúÜ2, th√¨ v√πng pixel ƒë√≥ l√† g√≥c.</li>
</ul>
n$
d{bmatrix}

<h4>Harris Corner trong OpenCV</h4>

```python
# cv2.cornerHarris(img_src, blockSize , kSize, k)
# --> img_src: ·∫£nh ƒë·∫ßu v√†o ph·∫£i l√† ·∫£nh m·ª©c x√°m (gray) v√† ·ªü ki·ªÉu float32
# --> blockSize: k√≠ch th∆∞·ªõc c·ªßa c·ª≠a s·ªï tr∆∞·ª£t tr√™n ·∫£nh
# --> kSize: k√≠ch th∆∞·ªõc c·ªßa h·∫°t nh√¢n Sobel ƒë·ªÉ t√≠nh gradient
# --> k: tham s·ªë trong c√¥ng th·ª©c Harris (k l√† h·∫±ng s·ªë l·∫•y trong ƒëo·∫°n [0.04, 0.06]

import cv2
import numpy as np
import matplotlib.pyplot as plt

# ƒê·ªçc ·∫£nh t·ª´ file
image = cv2.imread('harris_corner.png') 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Chuy·ªÉn ƒë·ªïi ·∫£nh sang ki·ªÉu float32
gray = np.float32(gray)

# √Åp d·ª•ng Harris Corner Detection
harris = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Ph√≥ng to c√°c k·∫øt qu·∫£ ƒë·ªÉ quan s√°t
harris = cv2.dilate(harris, None)

# ƒê·∫∑t ng∆∞·ª°ng ƒë·ªÉ ph√°t hi·ªán c√°c g√≥c
image[harris > 0.01 * harris.max()] = [0, 0, 255]

# Hi·ªÉn th·ªã k·∫øt qu·∫£
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Harris Corners')
plt.show()
```

<h4>Corner with SubPixel Accuracy</h4>

Corner detection with sub-pixel accuracy c·∫£i thi·ªán vi·ªác x√°c ƒë·ªãnh v·ªã tr√≠ c√°c g√≥c b·∫±ng c√°ch tinh ch·ªânh v·ªã tr√≠ g√≥c t·ª´ ƒë·ªô ph√¢n gi·∫£i c·ªßa ·∫£nh ban ƒë·∫ßu ƒë·∫øn m·ª©c ƒë·ªô ch√≠nh x√°c d∆∞·ªõi m·ªôt pixel. 

Nh∆∞ th∆∞·ªùng l·ªá, tr∆∞·ªõc ti√™n ch√∫ng ta c·∫ßn t√¨m c√°c g√≥c Harris. Sau ƒë√≥, ch√∫ng ta chuy·ªÉn tr·ªçng t√¢m c·ªßa c√°c g√≥c n√†y (C√≥ th·ªÉ c√≥ m·ªôt lo·∫°t pixel ·ªü m·ªôt g√≥c, ch√∫ng ta l·∫•y tr·ªçng t√¢m c·ªßa ch√∫ng) ƒë·ªÉ tinh ch·ªânh ch√∫ng. Ch√∫ng  d·ª´ng n√≥ sau m·ªôt s·ªë l·∫ßn l·∫∑p ƒë∆∞·ª£c ch·ªâ ƒë·ªãnh ho·∫∑c ƒë·∫°t ƒë∆∞·ª£c ƒë·ªô ch√≠nh x√°c nh·∫•t ƒë·ªãnh, t√πy theo ƒëi·ªÅu ki·ªán n√†o x·∫£y ra tr∆∞·ªõc. Ch√∫ng ta c≈©ng c·∫ßn x√°c ƒë·ªãnh k√≠ch th∆∞·ªõc c·ªßa v√πng l√¢n c·∫≠n m√† n√≥ s·∫Ω t√¨m ki·∫øm c√°c g√≥c.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('harris_corner.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

# √Åp d·ª•ng Harris Corner Detection
harris = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
ret, harris = cv2.threshold(harris, 0.01 * harris.max(), 255, 0)
harris = np.uint8(harris)

# T√¨m c√°c ƒëi·ªÉm centroid (tr·ªçng t√¢m)
ret, labels, stats, centroids = cv2.connectedComponentsWithStats(harris)

# Ti√™u ch√≠ d·ª´ng v√† t·ªëi ∆∞u h√≥a (d·ª´ng sau 100 l·∫ßn l·∫∑p ho·∫∑c ƒë·ªô ch√≠nh x√°c epsilon = 0.001)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)

# Tinh ch·ªânh c√°c v·ªã tr√≠ g√≥c ƒë·∫øn ƒë·ªô ch√≠nh x√°c sub-pixel
corners = cv2.cornerSubPix(gray, np.float32(centroids), (5, 5), (-1, -1), criteria)

# Hi·ªÉn th·ªã k·∫øt qu·∫£
res = np.hstack((centroids, corners))
res = np.intp(res)
image[res[:, 1], res[:, 0]] = [0, 0, 255]
image[res[:, 3], res[:, 2]] = [0, 255, 0]

# Hi·ªÉn th·ªã ·∫£nh v·ªõi c√°c g√≥c ƒë√£ tinh ch·ªânh
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Corners with SubPixel Accuracy')
plt.show()
```

H√†m cv2.cornerSubPix trong OpenCV c√≥ c√°c tham s·ªë sau:
<ul>
    <li>image: ƒê√¢y l√† ·∫£nh ƒë·∫ßu v√†o, th∆∞·ªùng l√† m·ªôt ·∫£nh grayscale. H√†m s·∫Ω th·ª±c hi·ªán t√≠nh to√°n n·ªôi suy tr√™n ·∫£nh n√†y ƒë·ªÉ ƒëi·ªÅu ch·ªânh c√°c v·ªã tr√≠ g√≥c.</li>
    <li>corners: M·ªôt m·∫£ng numpy ch·ª©a c√°c t·ªça ƒë·ªô c·ªßa c√°c ƒëi·ªÉm g√≥c ban ƒë·∫ßu. ƒê√¢y l√† ƒë·∫ßu ra t·ª´ ph∆∞∆°ng ph√°p ph√°t hi·ªán g√≥c nh∆∞ Harris Corner Detection.</li>
    <li>winSize: K√≠ch th∆∞·ªõc c·ªßa c·ª≠a s·ªï l√¢n c·∫≠n ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ t√≠nh to√°n n·ªôi suy. Th∆∞·ªùng l√† m·ªôt tuple (width, height), v√≠ d·ª•: (5, 5) ho·∫∑c (3, 3).</li>
    <li>zeroZone: K√≠ch th∆∞·ªõc c·ªßa v√πng kh√¥ng ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ t√≠nh to√°n trung b√¨nh di ƒë·ªông (moving average). Th∆∞·ªùng l√† m·ªôt tuple (width, height). N·∫øu kh√¥ng mu·ªën s·ª≠ d·ª•ng, c√≥ th·ªÉ ƒë·∫∑t gi√° tr·ªã n√†y l√† (-1, -1).</li>
    <li>criteria: Ti√™u ch√≠ d·ª´ng v√† t·ªëi ∆∞u h√≥a ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ x√°c ƒë·ªãnh khi n√†o thu·∫≠t to√°n tinh ch·ªânh s·∫Ω d·ª´ng. Th∆∞·ªùng l√† m·ªôt tuple g·ªìm hai ti√™u ch√≠: s·ªë l·∫ßn l·∫∑p t·ªëi ƒëa v√† sai s·ªë t·ªëi thi·ªÉu.</li>
</ul>

H√†m cv2.cornerSubPix trong OpenCV th·ª±c hi·ªán t√≠nh to√°n n·ªôi suy (interpolation) ƒë·ªÉ ƒëi·ªÅu ch·ªânh c√°c v·ªã tr√≠ g√≥c ƒë√£ ƒë∆∞·ª£c ph√°t hi·ªán t·ª´ Harris Corner Detection ho·∫∑c c√°c ph∆∞∆°ng ph√°p ph√°t hi·ªán g√≥c kh√°c ƒë·∫øn ƒë·ªô ch√≠nh x√°c sub-pixel. D∆∞·ªõi ƒë√¢y l√† c√°ch ho·∫°t ƒë·ªông c·ªßa h√†m n√†y:
<ul>
    <li> <b>  X√°c ƒê·ªãnh V√πng L√¢n C·∫≠n (Neighborhood): </b> ƒê·∫ßu ti√™n, h√†m n√†y x√°c ƒë·ªãnh m·ªôt v√πng l√¢n c·∫≠n xung quanh m·ªói ƒëi·ªÉm g√≥c. V√πng l√¢n c·∫≠n n√†y th∆∞·ªùng l√† m·ªôt c·ª≠a s·ªï ho·∫∑c m·ªôt v√πng nh·ªè xung quanh ƒëi·ªÉm g√≥c. </li>
    <li> <b>  N·ªôi Suy (Interpolation): </b> Sau khi x√°c ƒë·ªãnh v√πng l√¢n c·∫≠n, h√†m s·ª≠ d·ª•ng c√°c ph∆∞∆°ng ph√°p n·ªôi suy ƒë·ªÉ t√≠nh to√°n l·∫°i v·ªã tr√≠ ch√≠nh x√°c c·ªßa c√°c ƒëi·ªÉm g√≥c trong v√πng l√¢n c·∫≠n n√†y. C√°c ph∆∞∆°ng ph√°p n·ªôi suy th∆∞·ªùng ƒë∆∞·ª£c s·ª≠ d·ª•ng bao g·ªìm n·ªôi suy tuy·∫øn t√≠nh, bilinear interpolation, ho·∫∑c c√°c ph∆∞∆°ng ph√°p n·ªôi suy kh√°c. </li>
    <li> <b>  C·∫≠p Nh·∫≠t V·ªã Tr√≠ G√≥c: </b> Sau khi t√≠nh to√°n n·ªôi suy, h√†m c·∫≠p nh·∫≠t v·ªã tr√≠ c·ªßa c√°c ƒëi·ªÉm g√≥c d·ª±a tr√™n c√°c gi√° tr·ªã n·ªôi suy ƒë√£ ƒë∆∞·ª£c t√≠nh to√°n. </li>
    <li> <b>  Ti√™u Ch√≠ D·ª´ng (Stopping Criterion): </b> H√†m c√≥ th·ªÉ s·ª≠ d·ª•ng m·ªôt ti√™u ch√≠ d·ª´ng ƒë·ªÉ x√°c ƒë·ªãnh khi n√†o qu√° tr√¨nh t√≠nh to√°n n·ªôi suy s·∫Ω k·∫øt th√∫c. Ti√™u ch√≠ d·ª´ng n√†y th∆∞·ªùng li√™n quan ƒë·∫øn vi·ªác ƒë·∫°t ƒë·∫øn m·ªôt ng∆∞·ª°ng sai s·ªë nh·∫•t ƒë·ªãnh ho·∫∑c s·ªë l·∫ßn l·∫∑p t·ªëi ƒëa. </li>
    <li> <b> D√¢du Ra: </b> K·∫øt qu·∫£ tr·∫£ v·ªÅ c·ªßa h√†m l√† c√°c v·ªã tr√≠ g√≥c ƒë√£ ƒë∆∞·ª£c tinh ch·ªânh ƒë·∫øn ƒë·ªô ch√≠nh x√°c sub-pixel sau khi th·ª±c hi·ªán t√≠nh to√°n n·ªôi suy. </li>
</ul>

<h3>Shi-Tomasi Corner Detector</h3>

<h4>Shi-Tomasi Corner Detector</h4>

N·∫øu nh∆∞ Harris Corner Detector s·ª≠ d·ª•ng corner score function l√† $\mathrm{R} =  \mathrm{det(M)} - k(\mathrm{trace(M)})^2\hspace{1cm}$ th√¨ Shi-Tomasi Corner Detector s·ª≠ d·ª•ng h√†m t√≠nh ƒëi·ªÉm kh√°c l√† $R = min(\lambda _{1}, \lambda _{2})$. 

<ul>
    <li><b>image:</b>ƒê√¢y l√† ·∫£nh ƒë·∫ßu v√†o, th∆∞·ªùng l√† ·∫£nh grayscale.</li>
    <li><b>maxCorners:</b>S·ªë l∆∞·ª£ng t·ªëi ƒëa c√°c ƒëi·ªÉm g√≥c c·∫ßn ph√°t hi·ªán, c√°c ƒëi·ªÉm g√≥c t·ªët nh·∫•t (R max)</li>
    <li><b>qualityLevel:</b>ƒê√¢y l√† ng∆∞·ª°ng ch·∫•t l∆∞·ª£ng ƒë·ªÉ l·ª±a ch·ªçn c√°c ƒëi·ªÉm g√≥c. N√≥ l√† m·ªôt gi√° tr·ªã n·∫±m trong kho·∫£ng t·ª´ 0 ƒë·∫øn 1, ch·ªâ ra ch·∫•t l∆∞·ª£ng t·ªëi thi·ªÉu c·ªßa c√°c ƒëi·ªÉm g√≥c. C√°c ƒëi·ªÉm g√≥c c√≥ gi√° tr·ªã score ùëÖ cao h∆°n ng∆∞·ª°ng n√†y s·∫Ω ƒë∆∞·ª£c ch·ªçn.</li>
    <li><b>minDistance:</b>Kho·∫£ng c√°ch t·ªëi thi·ªÉu gi·ªØa hai ƒëi·ªÉm g√≥c</li>
    <li><b>mask:</b>m·∫£ng boolean ch·ªâ ƒë·ªãnh v√πng c·∫ßn quan t√¢m</li>
    <li><b>blockSize:</b>K√≠ch th∆∞·ªõc c·ª≠a s·ªï l√¢n c·∫≠n t√≠nh ƒë·∫°o h√†m</li>
    <li><b>useHarrisDetector:</b> true n·∫øu d√πng Harris. M·∫∑c ƒë·ªãnh l√† false cho Shi-Tomasi</li>
    <li><b>k:</b>Tham s·ªë k trong thu·∫≠t to√°n Harris Corner Detector</li>
</ul>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

original = cv2.imread('harris_corner.png')
img = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

# Ph√°t hi·ªán c√°c ƒëi·ªÉm g√≥c b·∫±ng Shi-Tomasi Corner Detector
corners = cv2.goodFeaturesToTrack(img, maxCorners=100, qualityLevel=0.1, minDistance=10)

# V·∫Ω c√°c ƒëi·ªÉm g√≥c l√™n ·∫£nh
corners = np.intp(corners)
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(original, (x, y), 1, 255, -1)

# Hi·ªÉn th·ªã ·∫£nh v·ªõi c√°c ƒëi·ªÉm g√≥c ƒë√£ ph√°t hi·ªán
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.title('Shi-Tomasi Corner Detector')
plt.show()
```

<h5>Non-maximal Suppression</h5>

ƒê·ªÉ gi·∫£i quy·∫øt v·∫•n ƒë·ªÅ c√°c ƒëi·ªÉm feature ƒë∆∞·ª£c x√°c ƒë·ªãnh c√≥ v·ªã tr√≠ li·ªÅn k·ªÅ nhau. 
<ul>
    <li>T√≠nh h√†m ƒëi·ªÉm V cho t·∫•t c·∫£ c√°c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng ƒë∆∞·ª£c ph√°t hi·ªán. V l√† t·ªïng ch√™nh l·ªách tuy·ªát ƒë·ªëi gi·ªØa p v√† 16 gi√° tr·ªã pixel xung quanh.</li>
    <li>S·∫Øp x·∫øp c·∫£c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng theo gi√° tr·ªã ƒëi·ªÉm V</li>
    <li>B·∫Øt ƒë·∫ßu t·ª´ ƒëi·ªÉm ƒë·∫∑c tr∆∞ng c√≥ ƒëi·ªÉm cao nh·∫•t, b·ªè ƒëi c√°c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng li·ªÅn k·ªÅ trong v√πng l√¢n c·∫≠n</li>

<h5>FAST trong OpenCV</h5>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('harris_corner.png', cv2.IMREAD_GRAYSCALE)

# T·∫°o ƒë·ªëi t∆∞·ª£ng FAST
threshold = 8
nonmax_suppression = True
fast = cv2.FastFeatureDetector_create(threshold, nonmax_suppression, cv2.FAST_FEATURE_DETECTOR_TYPE_9_16) 
# ƒë√¢y ƒëang d√πng lo·∫°i 9/16 (9 pixels trong 16 pixels l√¢n c·∫≠n), ngo√†i ra c√≤n c√≥ _5_8 v√† _7_12

# Ph√°t hi·ªán c√°c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng
keypoints = fast.detect(image, None)  # None l√† m·∫∑t n·∫° ·∫£nh (mask)

# V·∫Ω c√°c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng tr√™n ·∫£nh
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, color=(255, 0, 0))

# Hi·ªÉn th·ªã ·∫£nh
plt.figure(figsize=(10, 10))
plt.imshow(image_with_keypoints, cmap='gray')
plt.title('FAST Keypoints')
plt.show()

# In ra s·ªë l∆∞·ª£ng c√°c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng
print("Number of keypoints detected: ", len(keypoints))
```
