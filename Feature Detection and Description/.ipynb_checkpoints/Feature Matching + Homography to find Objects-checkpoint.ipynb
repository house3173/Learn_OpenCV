<h3>Feature Matching</h3>

<h4>Brute-Force Matcher</h4>

Brute-Force Matcherlấy mô tả của một tính năng trong tập đầu tiên và được so khớp với tất cả các tính năng khác trong tập thứ hai bằng cách sử dụng một số phép tính khoảng cách. Và tính năng có khoảng cách nhỏ nhất sẽ được trả lại.

Trước tiên, tạo đối tượng BFMatcher bằng cách sử dụng cv2.BFMatcher(). Cần chọn hai tham số: Đầu tiên là normType. Nó chỉ định phép đo khoảng cách sẽ được sử dụng. Theo mặc định là cv2.NORM_L2 (nó tốt cho SIFT, SURF, v.v. , cv2.NORM_L1cũng có ở đó). Đối với các bộ mô tả dựa trên chuỗi nhị phân như ORB, BRIEF, v.v., cv2.NORM_HAMMING nên sử dụng. Nếu ORB đang sử dụng , nên sử dụng VTA_K == 3 or 4 kết hợp cv2.NORM_HAMMING2.

Tham số thứ hai là biến boolean, crossCheck theo mặc định là sai. Nếu đúng, Matcher chỉ trả về những kết quả khớp có giá trị (i,j) sao cho bộ mô tả thứ i trong tập hợp A có bộ mô tả thứ j trong tập hợp B là kết quả phù hợp nhất và ngược lại. Nghĩa là, hai đặc điểm trong cả hai bộ phải khớp với nhau. Nó cung cấp kết quả nhất quán.

```python
import cv2
import matplotlib.pyplot as plt

# Đọc hai ảnh đầu vào
image1_path = 'stop_1.jpg'
image2_path = 'stop_2.jpg'

image1 = cv2.imread(image1_path)
image2 = cv2.imread(image2_path)

gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Khởi tạo đối tượng ORB
orb = cv2.ORB_create(nfeatures=1000, edgeThreshold=31)

# Tìm điểm đặc trưng và mô tả với ORB cho cả hai ảnh
keypoints1, descriptors1 = orb.detectAndCompute(gray1, None)
keypoints2, descriptors2 = orb.detectAndCompute(gray2, None)

# Sử dụng BFMatcher để so khớp các mô tả
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Tìm các cặp khớp tốt nhất
matches = bf.match(descriptors1, descriptors2)
matches = sorted(matches, key=lambda x: x.distance)

# Vẽ các cặp khớp tốt nhất
image_matches = cv2.drawMatches(image1, keypoints1, image2, keypoints2, matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Hiển thị ảnh với các cặp khớp sử dụng matplotlib
plt.figure(figsize=(15, 10))
plt.imshow(cv2.cvtColor(image_matches, cv2.COLOR_BGR2RGB))
plt.title('ORB Feature Matching')
plt.axis('off')
plt.show()

# In ra số lượng cặp khớp tìm thấy
print(f"Number of matches found: {len(matches)}")
```

```python
# sử dụng knnMatch(descriptor1, descriptor2, k = 2) để lấy hai đặc trưng tốt nhất ở image2 cho mỗi mô tả của 1 đặc trưng ở image1
# Tỷ lệ Lowe: Tìm k điểm khớp gần nhất (chằng hạn k = 2). Chỉ giữ lại điểm khớp gần nhất (m) nếu khoảng cách của nó nhỏ hơn 
# một tỷ lệ nhất định của khoảng cách tới điểm khớp gần thứ hai (n). (Thường là 0.75)

import cv2
import matplotlib.pyplot as plt

# Đọc hai ảnh đầu vào
image1_path = 'stop_1.jpg'
image2_path = 'stop_2.jpg'

image1 = cv2.imread(image1_path)
image2 = cv2.imread(image2_path)

gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Khởi tạo đối tượng ORB
orb = cv2.ORB_create()

# Tìm điểm đặc trưng và mô tả với ORB cho cả hai ảnh
keypoints1, descriptors1 = orb.detectAndCompute(gray1, None)
keypoints2, descriptors2 = orb.detectAndCompute(gray2, None)

# Sử dụng BFMatcher để so khớp các mô tả
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=False)

# Tìm 2 điểm khớp gần nhất cho mỗi mô tả
matches = bf.knnMatch(descriptors1, descriptors2, k=2)

# Áp dụng tỷ lệ Lowe để lọc các cặp khớp tốt nhất
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

# Vẽ các cặp khớp tốt nhất
image_matches = cv2.drawMatches(image1, keypoints1, image2, keypoints2, good_matches[:100], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Hiển thị ảnh với các cặp khớp sử dụng matplotlib
plt.figure(figsize=(15, 10))
plt.imshow(cv2.cvtColor(image_matches, cv2.COLOR_BGR2RGB))
plt.title('ORB Feature Matching with k=2 and Lowe\'s Ratio Test')
plt.axis('off')
plt.show()

# In ra số lượng cặp khớp tìm thấy
print(f"Number of good matches found: {len(good_matches)}")
```

<h4>FLANN based Matcher</h4>

FLANN là viết tắt của Fast Library for Approximate Nearest Neighbors. Nó chứa một tập hợp các thuật toán được tối ưu hóa để tìm kiếm hàng xóm gần nhất nhanh chóng trong các bộ dữ liệu lớn và cho các tính năng chiều cao. Nó hoạt động nhanh hơn BFMatcher đối với các tập dữ liệu lớn.

Để sử dụng FLANN thì đối với SIFT hay SURF, ta sử dụng <i>index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)</i>. Đối với OCR, ta sử dụng <i>index_params= dict(algorithm = FLANN_INDEX_LSH,  table_number = 6   key_size = 1     multi_probe_level =</i>

 thứ hai là SearchParams. Nó chỉ định số lần các cây trong chỉ mục phải được duyệt đệ quy. Giá trị cao hơn mang lại độ chính xác cao hơn nhưng cũng mất nhiều thời gian hơn. Nếu muốn thay đổi giá trị, hãy thay đổi search_params = dict(checks=100) #2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Đọc hai ảnh đầu vào
image1_path = 'stop_1.jpg'
image2_path = 'stop_2.jpg'

image1 = cv2.imread(image1_path)
image2 = cv2.imread(image2_path)

gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Khởi tạo đối tượng ORB
orb = cv2.ORB_create()

# Tìm điểm đặc trưng và mô tả với ORB cho cả hai ảnh
keypoints1, descriptors1 = orb.detectAndCompute(gray1, None)
keypoints2, descriptors2 = orb.detectAndCompute(gray2, None)

# Định nghĩa các tham số FLANN
FLANN_INDEX_LSH = 6
index_params = dict(algorithm = FLANN_INDEX_LSH,
                    table_number = 6,
                    key_size = 12,
                    multi_probe_level = 1)
search_params = dict(checks=50)

# Khởi tạo FLANN-based matcher
flann = cv2.FlannBasedMatcher(index_params, search_params)

# Tìm 2 điểm khớp gần nhất cho mỗi mô tả
matches = flann.knnMatch(descriptors1, descriptors2, k=2)

# Áp dụng tỷ lệ Lowe để lọc các cặp khớp tốt nhất
good_matches = []
for match in matches:
    if len(match) == 2:
        m, n = match
        if m.distance < 0.75 * n.distance:
            good_matches.append(m)
# Vẽ các cặp khớp tốt nhất
image_matches = cv2.drawMatches(image1, keypoints1, image2, keypoints2, good_matches[:100], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Hiển thị ảnh với các cặp khớp sử dụng matplotlib
plt.figure(figsize=(15, 10))
plt.imshow(cv2.cvtColor(image_matches, cv2.COLOR_BGR2RGB))
plt.title('ORB Feature Matching with FLANN-based Matcher')
plt.axis('off')
plt.show()

# In ra số lượng cặp khớp tìm thấy
print(f"Number of good matches found: {len(good_matches)}")
```

<h3>Feature Matching + Homography to find Objects</h3>

Thông qua phương pháp đối sánh, chúng ta thu được một số tập cặp điểm kết nối từ một queryImage (hay templateImage) đến một trainImage thể hiện một số bộ phận của vật thể trong ảnh queryImage trong một bức ảnh lộn xộn chứa vật thể đó là trainImage. Và thông tin này là đủ để tìm thấy vật thể trong bức ảnh lớn.

Sau khi thực hiện đối sánh hai bức ảnh, ta thu được danh sách các kết nối được cho là tốt. Để xác định vật thể, ta cần phải có đủ MIN_MATCH_COUNT cặp kết nối tốt. Nếu như không đủ thì hiển thị thông báo là chưa đủ dữ liệu để tìm kiếm. 

Khi đã có đủ tập hợp các điểm, sử dụng hàm <b>cv2.findHomography</b> trong thư viện OpenCV để tính toán phép biến đổi hình học đồng nhất (homography) giữa hai tập hợp các điểm. cv2.findHomography được sử dụng để tìm ma trận homography 𝐻 sao cho:<br>
$s\begin{bmatrix}x'\\ y'\\ 1\end{bmatrix} = H\begin{bmatrix}x\\ y\\ 1\end{bmatrix}$ <br>
Dễ nhận thấy H là một ma trận 3x3 thực hiện biến đổi điểm (x,y) thành điểm (x',y').

Hàm <b>cv2.findHomography</b> có các tham số gồm:
<ul>
    <li>srcPoints (tập hợp điểm đầu vào)</li> 
    <li>dstPoints (tập hợp điển kết quả)</li>
    <li>method (phương pháp ước lượng homography). 
        <ul>
            <li>0 hoặc cv2.RANSAC: Sử dụng thuật toán RANSAC (Random Sample Consensus).</li>
            <li>cv2.LMEDS: Sử dụng thuật toán Least-Median robust method.</li>
        </ul>
        Nếu sử dụng RANSAC, hàm sẽ lọc ra các inlier (các điểm phù hợp với mô hình homography) và trả về một mask cho biết điểm nào là inlier.  Hàm trả về ma trận homography H và mask
    </li>
    <li>ransacReprojThreshold: Ngưỡng tái chiếu (reprojection threshold) cho thuật toán RANSAC. Đây là khoảng cách tối đa mà một điểm tái chiếu (reprojected point) được xem là inlier. </li>
</ul>

Cuối cùng, ta có thể sử dụng cv2.perspectiveTransform() để tìm đối tượng. Nó cần ít nhất bốn điểm chính xác để tìm ra sự biến đổi.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

MIN_MATCH_COUNT = 10

img1 = cv2.imread('stop_1.jpg',0)          # queryImage
img2 = cv2.imread('stop_2.jpg',0)          # trainImage

# Initiate SIFT detector
sift = cv2.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)

FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)

flann = cv2.FlannBasedMatcher(index_params, search_params)

matches = flann.knnMatch(des1, des2, k=2)

# store all the good matches as per Lowe's ratio test.
good = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:
        good.append(m)

if len(good) > MIN_MATCH_COUNT:
    src_pts = np.float32([kp1[m.queryIdx].pt for m in good]).reshape(-1, 1, 2)
    dst_pts = np.float32([kp2[m.trainIdx].pt for m in good]).reshape(-1, 1, 2)

    # Find homography
    H, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)
    matchesMask = mask.ravel().tolist()

    h, w = img1.shape
    pts = np.float32([[0, 0], [0, h-1], [w-1, h-1], [w-1, 0]]).reshape(-1, 1, 2)
    dst = cv2.perspectiveTransform(pts, H)

    img2 = cv2.polylines(img2, [np.int32(dst)], True, 255, 3, cv2.LINE_AA)

    draw_params = dict(matchColor=(0, 255, 0), 
                       singlePointColor=None,
                       matchesMask=matchesMask, 
                       flags=2)

    img3 = cv2.drawMatches(img1, kp1, img2, kp2, good, None, **draw_params)

    plt.imshow(img3, 'gray')
    plt.show()

else:
    print("Not enough matches are found - {}/{}".format(len(good), MIN_MATCH_COUNT))
    matchesMask = None
```
