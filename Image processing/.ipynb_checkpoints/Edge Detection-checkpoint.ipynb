<h3>Sobel Derivatives</h3>

Hàm cv2.Sobel() trong thư viện OpenCV được sử dụng để tính toán đạo hàm bậc nhất của một hình ảnh theo hướng x hoặc y, sử dụng toán tử Sobel. Toán tử Sobel là một bộ lọc vi phân, nó giúp làm nổi bật các cạnh trong hình ảnh bằng cách tính toán gradient của cường độ ảnh tại mỗi điểm pixel.

<div style="padding: 20px; background-color: gray; border: 1px solid black;">
    <strong style="color: white;">cv2.Sobel(src, ddepth, dx, dy, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)</strong>
</div>

<ul>Trong đó:
<li>src: Ảnh đầu vào (ảnh xám hoặc ảnh màu).</li>
<li>ddepth: Độ sâu của ảnh đích (thường sử dụng cv2.CV_64F để giữ lại giá trị gradient có thể là số âm).</li>
<li>dx: Bậc đạo hàm theo hướng x.</li>
<li>dy: Bậc đạo hàm theo hướng y.</li>
<li>ksize: Kích thước của kernel Sobel. Giá trị mặc định là 3.</li>
<li>scale: Hệ số tỷ lệ cho giá trị đạo hàm. Giá trị mặc định là 1.</li>
<li>delta: Giá trị cộng thêm vào kết quả đạo hàm. Giá trị mặc định là 0.</li>
<li>borderType: Kiểu biên được sử dụng. Giá trị mặc định là cv2.BORDER_DEFAULT.</li>
</ul>

```python
Cách thức hoạt động
1. Tính toán đạo hàm:
Đối với mỗi pixel trong hình ảnh, hàm cv2.Sobel() sẽ tính toán gradient theo hướng x và/hoặc y bằng cách sử dụng kernel Sobel.
 --> Nếu dx=1 và dy=0, hàm sẽ tính toán đạo hàm bậc nhất theo hướng x.
 --> Nếu dx=0 và dy=1, hàm sẽ tính toán đạo hàm bậc nhất theo hướng y.
 --> Nếu cả dx và dy đều lớn hơn 0, hàm sẽ tính toán đạo hàm hỗn hợp.

2. Kết hợp kết quả đạo hàm:
 --> Kết quả của đạo hàm sẽ được nhân với hệ số scale và cộng với delta.
 --> Kết quả cuối cùng sẽ được lưu trữ trong biến đích với độ sâu xác định bởi ddepth.
```

```python
Để tính đạo hàm (lấy ví dụ với kernel_size = 3):
--> Kernel Sobel cho hướng X:
Gx = [[-1, 0, 1],
      [-2, 0, 2],
      [-1, 0, 1]]
--> Kernel Sobel cho hướng y:
Gy = [[-1, -2, -1],
      [ 0,  0,  0],
      [ 1,  2,  1]]

Khi đó đạo hàm của pixel (x,y) có cường độ I(x,y) theo các trục X, Y sẽ là một phép nhân tích chập của kernel và ma trận với pixel (x,y) là trung tâm
Tổng hợp kết quả gradient theo cả hai hướng X, Y có thể dùng G(x,y) = sqrt(Ix^2 + Iy^2) hoặc G(x,y) = |Ix| + |Iy|
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('edge_detection.jpg', cv2.IMREAD_GRAYSCALE)

# Tính toán gradient theo hướng X
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

# Tính toán gradient theo hướng Y
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Tính độ lớn gradient tổng hợp
gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)

# Chuyển đổi kết quả về kiểu uint8 để hiển thị
abs_sobelx = np.absolute(sobelx)
abs_sobely = np.absolute(sobely)
sobelx_8u = np.uint8(abs_sobelx)
sobely_8u = np.uint8(abs_sobely)
gradient_magnitude_8u = np.uint8(gradient_magnitude)

plt.subplot(2,2,1),plt.imshow(img, cmap = 'gray')
plt.title('Original'), plt.axis('off') 
plt.subplot(2,2,2),plt.imshow(sobelx_8u, cmap = 'gray')
plt.title('Sobel X'), plt.axis('off') 
plt.subplot(2,2,3),plt.imshow(sobely_8u, cmap = 'gray')
plt.title('Sobel Y'), plt.axis('off') 
plt.subplot(2,2,4),plt.imshow(gradient_magnitude_8u, cmap = 'gray')
plt.title('Sobel XY'), plt.axis('off') 

plt.show()
```

<h3>Laplacian Derivatives</h3>

```python
Hàm cv2.Laplacian() trong thư viện OpenCV được sử dụng để tính toán đạo hàm bậc hai của một hình ảnh. 

Đối với mỗi pixel trong hình ảnh, hàm cv2.Laplacian() sẽ tính toán đạo hàm bậc hai bằng cách sử dụng kernel Laplacian.
    
Kernel Laplacian là một ma trận xác định, ví dụ kernel 3x3 chuẩn có thể là:
[[0,  1, 0],
 [1, -4, 1],
 [0,  1, 0]]

Hoặc

[[1,  1, 1],
 [1, -8, 1],
 [1,  1, 1]]
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Chuyển đổi kết quả về kiểu uint8 để hiển thị
abs_laplacian = np.absolute(laplacian)
laplacian_8u = np.uint8(abs_laplacian)

plt.imshow(laplacian_8u, cmap = 'gray')
plt.show()
```

<h3>Canny edge detection</h3>

<h4>1. Noise Reduction</h4h4>

Vì việc phát hiện cạnh dễ bị ảnh hưởng bởi nhiễu trong ảnh nên bước đầu tiên chúng ta cần giảm nhiễu bằng cách sử dụng bộ lọc Gauss với kernel = 5

<h4>2. Finding Intensity Gradient of the Image</h4>

Hình ảnh đã làm mịn được lọc bằng hạt nhân Sobel theo cả hướng ngang và hướng dọc để lấy đạo hàm bậc nhất theo hướng ngang ( Gx) và hướng dọc ( Gy). Từ hai hình ảnh này, chúng ta có thể tìm độ dốc cạnh và hướng cho từng pixel (Hướng gradient luôn vuông góc với các cạnh) như sau: <br>
<img src="https://opencv24-python-tutorials.readthedocs.io/en/latest/_images/math/fc9752466c9c38d07985d62e86946489e23c61e2.png"><br>

<h4>3. Non-maximum Suppression</h4>

```python
Để có được các cạnh chính xác, thuật toán Canny loại bỏ các pixel không phải là cực đại trong hướng gradient. 
Điều này đảm bảo rằng chỉ có các điểm cường độ cực đại trên gradient được giữ lại.

--> Góc gradient được lượng tử hóa thành bốn hướng: 0°, 45°, 90°, và 135°.
--> Đối với mỗi pixel, kiểm tra xem giá trị gradient của nó có lớn hơn giá trị của hai pixel liền kề trong hướng gradient hay không. 
Nếu không, giá trị của pixel đó sẽ bị đặt về 0.
```

<h4>4. Ngưỡng kép (Double Threshold)</h4>

```python
Các pixel còn lại sau bước Non-maximum Suppression được phân loại thành ba loại 
dựa trên hai ngưỡng: ngưỡng cao (highThreshold) và ngưỡng thấp (lowThreshold).

--> Các pixel có giá trị gradient lớn hơn ngưỡng cao được coi là các biên cạnh mạnh.
--> Các pixel có giá trị gradient nằm giữa ngưỡng cao và ngưỡng thấp được coi là các biên cạnh yếu.
--> Các pixel có giá trị gradient nhỏ hơn ngưỡng thấp bị loại bỏ.
```

<h4>5. Kết nối biên cạnh (Edge Tracking by Hysteresis)</h4>

```python
Biên cạnh yếu chỉ được giữ lại nếu chúng kết nối với biên cạnh mạnh, tạo thành các biên cạnh cuối cùng. 
Điều này giúp loại bỏ các biên cạnh yếu do nhiễu và giữ lại các biên cạnh liên tục.

--> Tạo một ảnh nhị phân mới, bắt đầu từ các biên cạnh mạnh.
--> Đối với mỗi biên cạnh mạnh, kiểm tra các pixel lân cận.
--> Nếu có một pixel lân cận là biên cạnh yếu, nâng cấp pixel đó thành biên cạnh mạnh và tiếp tục kiểm tra các pixel lân cận của nó.
--> Lặp lại cho đến khi không còn pixel nào được nâng cấp.
```

```python
OpenCV puts all the above in single function, cv2.Canny(). We will see how to use it. 
    cv2.Canny(src, minVal, maxVal, kernel_Sobel_size, (bool)L2gradient)
    
--> First argument is our input image. 
--> Second and third arguments are our minVal and maxVal respectively. 
--> Third argument is aperture_size. It is the size of Sobel kernel used for find image gradients. By default it is 3. 
--> Last argument is L2gradient which specifies the equation for finding gradient magnitude. 
If it is True, it uses the equation mentioned above which is more accurate, otherwise it uses this function: (G) = |G_x| + |G_y|. 
By default, it is False.
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_offical = cv2.imread('blue_image.jpg')
img = cv2.imread('blue_image.jpg', 0)
edges = cv2.Canny(img, 100, 200)

plt.subplot(131),plt.imshow(img_offical[:,:,::-1])
plt.title('Original Image'), plt.axis('off')
plt.subplot(132),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.axis('off')
plt.subplot(133),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.axis('off')

plt.show()
```
