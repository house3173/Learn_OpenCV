```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import felzenszwalb

# Đọc hình ảnh đầu vào
image = cv2.imread('water_coins.jpg')
# Chuyển đổi từ không gian màu BGR sang RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Áp dụng thuật toán Felzenszwalb cho phân đoạn hình ảnh
segments = felzenszwalb(image_rgb, scale=100, sigma=0.5, min_size=40)

# Hiển thị kết quả phân đoạn
plt.figure(figsize=(5, 5))
plt.imshow(segments, cmap='nipy_spectral')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import felzenszwalb
from skimage.color import rgb2gray
from skimage.measure import regionprops, label

# Đọc hình ảnh đầu vào
image = cv2.imread('water_coins.jpg')
# Chuyển đổi từ không gian màu BGR sang RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Áp dụng thuật toán Felzenszwalb cho phân đoạn hình ảnh với các tham số khác nhau
segments = felzenszwalb(image_rgb, scale=200, sigma=0.5, min_size=200)

# Chuyển đổi hình ảnh RGB sang grayscale
gray_image = rgb2gray(image_rgb)

# Tạo ảnh nhãn từ các phân đoạn
label_image = label(segments)

# Lấy các thuộc tính của từng vùng
regions = regionprops(label_image)

# Vẽ hình chữ nhật bao quanh từng vùng
for region in regions:
    min_row, min_col, max_row, max_col = region.bbox
    cv2.rectangle(image_rgb, (min_col, min_row), (max_col, max_row), (0, 255, 0), 2)

# Hiển thị kết quả
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
